```python
import matplotlib.pyplot as plt
import numpy as np

# 함수 정의
def f(x):
    return 2 * x + 1

# x 값 범위 지정 (-10부터 10까지 100개 점)
x = np.linspace(-10, 10, 100)
y = f(x)

# 그래프 그리기
plt.plot(x, y, label='f(x) = 2x + 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Example of function graph')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def is_injective(f, domain):
    image = set()
    for x in domain:
        fx = f(x)
        if fx in image:
            return False
        image.add(fx)
    return True

domain = range(-10, 11)
print("Is it injective func?", is_injective(f, domain))
```

```python
def inverse_f(y):
    return (y - 1) / 2

y_values = np.linspace(-10, 10, 100)
x_values = inverse_f(y_values)

plt.plot(y_values, x_values, label='Inverse func f^{-1}(y) = (y - 1)/2', color='red')
plt.xlabel('y')
plt.ylabel('f^{-1}(y)')
plt.title('Inverse function graph')
plt.legend()
plt.grid(True)
plt.show()
```
