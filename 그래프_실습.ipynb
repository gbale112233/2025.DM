```python
graph = {
    'Alice': ['Bob', 'Charlie'],
    'Bob': ['David', 'Eve'],
    'Charlie': ['Frank'],
    'David': [],
    'Eve': ['Frank'],
    'Frank': []
}
```

```python
from collections import deque

def recommend_friend(graph, start):
    visited = set()
    queue = deque([(start, 0)])  # (노드, 거리)

    print(f"\n{start}의 친구 추천 (2단계):")

    while queue:
        current, depth = queue.popleft()
        if current not in visited:
            visited.add(current)
            if depth == 2:
                print("-", current)
            if depth < 2:
                for neighbor in graph[current]:
                    queue.append((neighbor, depth + 1))

# 예시 그래프
graph = {
    'Alice': ['Bob', 'Charlie'],
    'Bob': ['David', 'Eve'],
    'Charlie': ['Frank'],
    'David': [],
    'Eve': ['Frank'],
    'Frank': []
}

# 실행
recommend_friend(graph, 'Alice')

```
